```python
import json
class Person:
    def __init__(self, name, species, gender, status, image):
        self.name = name
        self.species = species
        self.gender = gender
        self.status = status
        self.image = image
    @classmethod
    def from_json(cls, json_path):
        with open(json_path, "r") as f:
            data = json.load(f)
        return cls(data["name"], data["species"], data["gender"], data["status"], data["image"])
    
    def __repr__(self) :
        return f"Это {self.name}, статус: {self.status}"
# Использование
person = Person.from_json(f'{base_dir}/1.json')
print(person.name) 
print(person.species)  
print(person.gender)
print(person.status)
```

```python
import os
# Укажите путь к папке с JSON-файлами
folder_path = f'{base_dir}'
# Инициализируем пустой список для хранения данных
combined_data = []
# Проходим по всем файлам в указанной папке
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):  # Проверяем, что файл имеет расширение .json
        file_path = os.path.join(folder_path, filename)
        print(file_path)
        combined_data.append(Person.from_json(file_path))
# Теперь combined_data содержит данные из всех JSON-файлов
print(len(combined_data))  # Выводим количество объединенных записей
```
